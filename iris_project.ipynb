```python
x = iris.drop(['species'],axis = 1)
```

```python
x
```

```python
y = iris['species']
```

```python
y
```

```python
le = LabelEncoder()
y = le.fit_transform(y)
y
#setosa:0,versicolor:1,virginia:2
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.3,random_state = 1)
```

```python
dtree = DecisionTreeClassifier()
dtree.fit(x_train,y_train)
```

```python
y_pred = dtree.predict(x_test)
```

```python
#Logistic Regression
```

```python
lr = LogisticRegression(solver = 'newton-cg')
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred1 = lr.predict(x_test) 
```

```python
sns.heatmap(data = confusion_matrix(y_test,y_pred1), annot = True,cmap = 'Blues')
plt.ylabel = ('Actual label')
plt.ylabel = ('Predicted label')
```

```python
accuracy_score(y_test,y_pred1)
```

```python
#Decision Tree
```

```python
dtree = DecisionTreeClassifier()
dtree.fit(x_train,y_train)
```

```python
y_pred2 = dtree.predict(x_test) 
```

```python
confusion_matrix(y_test,y_pred2)
```

```python
sns.heatmap(data = confusion_matrix(y_test,y_pred2), annot = True,cmap = 'Blues')
plt.ylabel = ('Actual label')
plt.ylabel = ('Predicted label')
```

```python
accuracy_score(y_test,y_pred2)
```

```python
plt.figure(figsize=(10,10))
dec_tree = plot_tree(decision_tree = dtree,feature_names = iris.columns[:-1],class_names = ["setosa","vericolor","verginica"],filled = True)
```

```python
#Random Forest
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
clf = RandomForestClassifier(n_estimators = 2000)
```

```python
clf.fit(x_train,y_train)
```

```python
y_pred3 = clf.predict(x_test) 
```

```python
confusion_matrix(y_test,y_pred3)
```

```python
sns.heatmap(data = confusion_matrix(y_test,y_pred3), annot = True,cmap = 'Blues')
plt.ylabel = ('Actual label')
plt.ylabel = ('Predicted label')
```

```python
accuracy_score(y_test,y_pred3)
```
